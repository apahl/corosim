```python
import pandas as pd

import matplotlib.pyplot as plt

from IPython.core.display import HTML, display 

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import plt_style
plt.style.use("seaborn-white")
```

```python
input = open("data.txt").read()
inp_list = input.split("\n")
if len(inp_list[0]) < 10:
    inp_list = inp_list[1:]
    
title = inp_list[0]
inp_list = [list(map(int, x.replace(" ", "\t").strip("\t").split("\t"))) for x in inp_list[2:]]
# inp_list = [list(x.strip("\t").split("\t")) for x in input.split("\n") if len(x) > 3]

df = pd.DataFrame(inp_list, columns=["Time", "Total", "Normal", "Infected", "Healed"])
df.head()

df.plot.line(x="Time", y=["Total", "Normal", "Infected", "Healed"], color=["black", "blue", "red", "green"], title=title);
```
